```python
fare = pd.read_csv("I:\NVTA\Inputs\Transit_Fare.txt", sep='\t', skiprows=1)
fare.head(10)
```

```python
fare_metrorail = fare[(fare['FROM_MODE'] == 'RAPIDRAIL') & (fare['TO_MODE'] == 'RAPIDRAIL') & (fare['CLASS'] == 'CASH') & (fare['PERIOD'] == '6:00..9:00, 15:00..19:00')]
fare_metro = fare_metrorail[['FROM_ZONE','TO_ZONE','FARE']]
fare_metro.head()
```

```python
fare_metro = fare_metro.set_index(['FROM_ZONE', 'TO_ZONE'])
```

```python
fare_metro_matrix = fare_metro.unstack(level=1)
fare_metro_matrix.head(10)
```

```python
fare_metro_matrix.to_csv(r"D:\Projects\2016\NVTA\Working\Metro_Fare.csv", sep=',',header=True,index=True,index_label=False)
```

```python
fare_metro_matrix.reset_index().head(10)
```

```python
fare_metro_matrix.reset_index().to_csv(r"D:\Projects\2016\NVTA\Working\Metro_Fare.csv", sep=',',header=False,index=False,index_label=False)
```

```python
## OFF PEAK
fare_metrorail_off_peak = fare[(fare['FROM_MODE'] == 'RAPIDRAIL') & (fare['TO_MODE'] == 'RAPIDRAIL') & (fare['CLASS'] == 'CASH') & (fare['PERIOD'] != '6:00..9:00, 15:00..19:00')]
fare_metro_off_peak = fare_metrorail_off_peak[['FROM_ZONE','TO_ZONE','FARE']]
fare_metro_off_peak.head()
```

```python
fare_metro_off_peak = fare_metro_off_peak.set_index(['FROM_ZONE', 'TO_ZONE'])
fare_metro_matrix_off_peak = fare_metro_off_peak.unstack(level=1)
fare_metro_matrix_off_peak.head(10)
```

```python
fare_metro_matrix_off_peak.reset_index().to_csv(r"D:\Projects\2016\NVTA\Working\Metro_Fare_off_peak.csv", sep=',',header=True,index=False,index_label=False)
```

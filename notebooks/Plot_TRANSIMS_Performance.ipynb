```python
def get_delay_reduction(filename_delay, filename_volume, threshold_delay = 0, threshold_volume = 0):
    """
    filename_delay: delay difference file
    filename_vol  : volume difference file
    
    Accumulate the delay differences satisfying both delay and volume thresholds
    
    """
    if os.path.isfile(filename_delay) and os.path.isfile(filename_volume):
        delay_diff = pd.read_csv(filename_delay, sep=",")
        volume_diff = pd.read_csv(filename_volume, sep=",")
        diff = delay_diff.join(volume_diff.set_index(["LINK","DIR"]), on=("LINK","DIR"), how="inner", lsuffix="_delay", rsuffix="_volume")
        if threshold_delay < 0 or threshold_volume < 0:
            return sum(diff.D000_3000_delay)
        else:
            return sum(diff.D000_3000_delay[(abs(diff.D000_3000_delay / (diff.B000_3000_delay + 1) < -threshold_delay)) | 
                                            (abs(diff.D000_3000_volume) > threshold_volume)                                           
                                           ])
    else:
        return 0
```

```python
def get_delay_select_link(filename_delay, exclude_links):
    if os.path.isfile(filename_delay):
        delay_diff = pd.read_csv(filename_delay, sep="\t")
        delay_diff = delay_diff.loc[~delay_diff.LINK.isin(exclude_links)]
        return sum(delay_diff.VHDX)
    else:
        return 0
```

```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
```

```python
# $VEHICLE:SIMRUN;NO;VEHTYPE;PTLINE;PTSTOP;SIMSEC
table_def = {'$VEHICLE:SIMRUN': np.int16, 'NO': np.int64, 'VEHTYPE': np.int16, 'PTLINE': np.int16, 'PTSTOP': np.int16, 'SIMSEC': np.float16}
```

```python
df = pd.read_table(r'D:\Projects\2016\Python\Run_001.fzp',sep=';',skiprows=16)
df.head()
```

```python
df.dtypes
```

```python
# Drop all PTSTOP with NaN
df2 = df.dropna(axis=0)
df2.head()
```

```python
# Group-by select the min simsec and max simsec
df2_group = df2.groupby(['NO','PTLINE', 'PTSTOP'])
```

```python
df2_head = df2_group.head(1)
```

```python
df2_tail = df2_group.tail(1)
```

```python
# merge using ['NO','PTLINE', 'PTSTOP']
merged = pd.merge(df2_head, df2_tail, left_on = ['NO','PTLINE', 'PTSTOP'], right_on=['NO','PTLINE', 'PTSTOP'], suffixes=['_min','_max'])
```

```python
merged_sort = merged.sort_values(by=['NO','PTLINE', 'SIMSEC_min'])
merged_sort[merged_sort['NO']==4]
```

```python
merged_sort.head(50)
```

```python
# compute the travel time = Next Arrival Time - Prev. Departure Time
depart = (merged_sort['SIMSEC_max'][0:len(merged_sort)-1]).values
arrival = (merged_sort['SIMSEC_min'][1:]).values
```

```python
depart
```

```python
arrival
```

```python
tt = arrival - depart
```

```python
tt[0]
```

```python
tt2 = np.zeros(len(tt)+1)
tt2[1:] = tt
tt2
```

```python
merged_sort.drop('tt',axis=1, inplace=True)
```

```python
merged_sort['tt'] = tt2
```

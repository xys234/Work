### How many links are there in each network?

```python
print("Number of links in the Network WITHOUT double CC = {0:d}".format(network_without_double.shape[0]))
print("Number of links in the Network WITH    double CC = {0:d}".format(network_with_double.shape[0]))
```

#### is every link in the WITHOUT present in the WITH?

```python
AB_pairs_without = network_without_double.set_index(['A','B']).index
AB_pairs_with = network_with_double.set_index(['A','B']).index
```

```python
AB_pairs_without.difference(AB_pairs_with)
```

#### filter out all links in WITH but not in WITHOUT

```python
links_not_in_without = network_with_double.set_index(['A','B']).loc[AB_pairs_with.difference(AB_pairs_without)]
```

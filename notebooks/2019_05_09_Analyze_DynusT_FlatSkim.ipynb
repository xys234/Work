```python
DYNUST_SKIM_DIR = r'C:\Projects\SWIFT\SWIFT_Workspace\Scenarios\Scenario_S0\STM\STM_A\01_DynusT\03_Model'
DYNUST_SKIM_FILE = os.path.join(DYNUST_SKIM_DIR, 'Vehicle_Skim.csv')
DYNUST_SKIM_FILE
```

```python
skim = pd.read_csv(DYNUST_SKIM_FILE)
skim.head()
```

```python
skim.head(20)
```

```python
total_vmt, total_vhd, total_vht = \
np.sum(skim.dist.values), np.sum(skim.delay.values) / 60.0, np.sum(skim.end.values - skim.start.values) / 60.0
```

```python
print('Total VMT = {:,.0f}'.format(total_vmt))
print('Total VHT = {:,.0f}'.format(total_vht))
print('Total VHD = {:,.0f}'.format(total_vhd))
```
